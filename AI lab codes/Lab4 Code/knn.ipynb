```python
import numpy as np
#import matplotlib as mtp
#from turtle import color 
from matplotlib import pyplot as plt
from matplotlib import style
from collections import Counter
style.use('fivethirtyeight')
```

Taking a demo Dataset 

```python
dataset = {
    'k' : [[210,95],[200,85],[190,87],[220,86],[200,93],[210,94],[203,90],[203,93],[213,91],[221,89]],
    'r' : [[170,80],[168,82],[190,78],[170,79],[160,83],[150,74],[180,75],[190,70],[160,60],[150,68]],
}
```

Scatter Plot 

```python
def scatter_plot(new_plot):
    #Plot the data from the dataset
    for group in dataset:
        for features in dataset[group]: 
            plt.scatter(features[0],features[1], s=50,color=group)

    #Plot the input data
    plt.scatter(new_plot[0],new_plot[1],color="b",marker= '*',s=50)
    plt.show()
```

```python
scatter_plot([175,65])
```

KNN Function Creation

```python
def K_NN(data,predict,k):
    distances = []
    
    #Determining the distance
    for group in data:
        for feature in data[group]:
            #euclidian distance = root( (X2-X1)*2 - (Y2-Y1)*2 )
            euclidian_distance = np.linalg.norm(np.array(feature)-np.array(predict))
            #Now storing all the points distance from the input point 
            distances.append([euclidian_distance,group])
       
    #Takes K number of points for sortest distance      
    votes = [i[1] for i in sorted(distances)[:k]] 
     
    #Counts the Group name for the K number of points
    vote_result = Counter(votes).most_common(1)[0][0]
    
    #Return the group name having the most common value or the maximum value
    return vote_result
```

Taking input from user and creating solution

```python
while True:
    #Taking an input from the user as height,weight = 175,65
    user_input = input('Enter the new sample[height,weight]:').split(',')
    player_sample = []
    
    #Placing new_sample = height then we have new_sample[height]
    player_sample.append(int(user_input[0]))
    #Placing new_sample = weight then we have new_sample[height,weight]
    player_sample.append(int(user_input[1]))
    
    #If new_sample = [0,0]
    if player_sample == [0,0]:
        break
    #else new_sample will be send to predict the player type
    else:
        print(f"Given input was {player_sample}")
        scatter_plot(player_sample)
        result = K_NN(dataset,player_sample,5)
    
    if result == 'k':
        print("The player is a Kabadi player")
    else:
        print("The player is a Wrestler")
```
